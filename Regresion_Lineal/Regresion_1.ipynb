#Importamos la data

```python
import pandas as pd 
df = pd.read_excel("PROD.xlsx")
data = pd.DataFrame(df)
data1=data.drop(['Additive'],axis=1) 
data1.head()
```

```python
data1.info()
```

```python
data1.describe()
```

```python
data1.corr()
```

```python

x = data1["Mat3"].values
y = data1["Production"].values
x = x.reshape(-1,1)
x.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x, y)
```

```python
predictions = reg.predict(x)
predictions
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y, color='blue', marker='.')
plt.plot(x, predictions, color='g')
plt.xlabel('Material 3')
plt.ylabel('Production')
plt.show()
```
